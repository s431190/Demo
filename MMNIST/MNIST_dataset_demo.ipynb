```python
from PIL import Image
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

IMAGE_PATH = './MMNIST_dataset_demo/'
IMAGE_FORMAT = ('.PNG', '.png', '.jpg', '.JPG ','.tiff')
IMAGE_NUM = 60
IMAGE_ROW = 6
IMAGE_COL = 10
IMAGE_SIZE = (28, 28)
IMAGE_NAME = './MMNISR_dataset_demo.PNG'

imgs = [img for img in os.listdir(IMAGE_PATH) for format in IMAGE_FORMAT if format == os.path.splitext(img)[1]]
if len(imgs) > IMAGE_ROW * IMAGE_COL:
    print('Warning: overflow')
    imgs = imgs[:len(imgs) - 1]

finalImg = Image.new('RGBA', (IMAGE_COL * IMAGE_SIZE[0], IMAGE_ROW * IMAGE_SIZE[1]))

curCol = curRow = 0
for name in imgs:
    img = Image.open(IMAGE_PATH + name)
    img = Image.open(IMAGE_PATH + name).resize(IMAGE_SIZE)
    finalImg.paste(img, (curCol * IMAGE_SIZE[0], curRow * IMAGE_SIZE[1]))
    curCol = curCol + 1
    if curCol >= IMAGE_COL:
        curCol = 0
        curRow = curRow + 1

finalImg.save(IMAGE_NAME)
# finalImg.show()

im = mpimg.imread(IMAGE_NAME)
imgplot = plt.imshow(im)
plt.show()
```
